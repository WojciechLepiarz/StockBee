# Google Trends analysis

```python
from pytrends.request import TrendReq
import pandas as pd
import pandas_bokeh
from bokeh.io import output_notebook
output_notebook()
```

```python
pytrends = TrendReq(hl='pl', tz=0)
#    Europe/Warsaw
kw_list = ["blockchain"]
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='PL', gprop='')
```

```python
df = pytrends.interest_over_time()
df.plot_bokeh()
```

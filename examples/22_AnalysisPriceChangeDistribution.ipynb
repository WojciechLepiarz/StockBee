```python
def getPriceChangeData(path: str, periods: list) -> tuple:
    sourceDataFrame = pd.read_csv(path, index_col=1)
    sourceDataFrame.index = pd.to_datetime(sourceDataFrame.index, format="%Y%m%d")
    labels = [f'Diff%_{str(period)}' for period in periods]
    for period in periods:
        sourceDataFrame[f'Diff%_{str(period)}'] = (sourceDataFrame['<CLOSE>'] / sourceDataFrame['<CLOSE>'].shift(period) - 1)*100 
    return sourceDataFrame[labels].dropna(how='any').reset_index(drop=True)

periods = [1, 5, 20, 60, 125, 255]
labels = [f'Diff%_{str(period)}' for period in periods]
data = pd.DataFrame([], columns=labels)
for symbol in gpw:
    data = data.append(getPriceChangeData(symbol, periods), ignore_index = True)
data
```
